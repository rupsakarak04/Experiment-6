```python
# 6.1) To check whether a number is positive, negative or zero
number = int(input("Enter a number: "))
if number > 0:
  print("The number is positive.")
elif number < 0:
  print("The number is negative.")
else:
  print("The number is zero.")
```

```python
# 6.4) Get input from user for a subject and calculate grade if-elif-else
marks = int(input("Enter your marks: "))
if marks >= 90:
  print("Grade: A")
elif marks >= 75:
  print("Grade: B")
elif marks >= 60:
  print("Grade: C")
elif marks >= 40:
  print("Grade : D")
else:
  print("Grade: Fail")
```

```python
# 6.5) Get input from user for 3 subjects, calculate average and calculate grade using if-elif-else
sub1= int(input("Enter your marks of first subject: "))
sub2 = int(input("Enter your marks of second subject: "))
sub3 = int(input("Enter your marks of third subject: "))
avg = (sub1 + sub2 + sub3) / 3
if avg >= 90:
  print("Grade: A")
elif avg >= 75:
  print("Grade: B")
elif avg >= 60:
  print("Grade: C")
elif avg >= 40:
  print("Grade : D")
else:
    print("Grade: F")
```

```python
# 6.6) Create a program to check whether the given year is leap year or not
year = int(input("Enter a year: "))
if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
  print("The year is a leap year.")
else:
   print("The year is not a leap year.")
```

```python
# 6.9) Calculating salary of a employee with allowances
basic_salary = float(input("Enter the basic salary: "))

if basic_salary <= 10000:
  HRA = 0.20 * basic_salary
  DA = 0.80 * basic_salary
elif basic_salary <= 20000:
  HRA = 0.25 * basic_salary
  DA = 0.90 * basic_salary
else:
  HRA = 0.30 * basic_salary
  DA = 0.95 * basic_salary

gross_salary = basic_salary + HRA + DA
print("Basic Salary:", basic_salary)
print("HRA:", HRA)
print("DA:", DA)
print("Gross Salary:", gross_salary)
```
